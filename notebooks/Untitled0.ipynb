```python
import os
import pickle
```

```python
data=[]
def get_all(pickled=False, get_columns_only=False, to_csv=False):
    data=[]
    all_data=[]
    labels=[]
    dir_name='saved_animations_and_data'
    sample_sets=os.listdir(dir_name)
    for sample_set in sample_sets:
        if sample_set[0]!= '.': 
            session_files= os.listdir(os.path.join(dir_name,sample_set))
            for f in session_files:
#                 print f
                if f[-2:]==".p":
                    data+=pickle.load(open(os.path.join(dir_name,sample_set,f)))
    return data
            
            
            
data=get_all()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
df.columns=['User','exerciseType','rep_count','t (sec)','acc_x','acc_y','acc_z','gyro_x','gyro_y','gyro_z','magnet_x','magnet_y','magnet_z']
```
